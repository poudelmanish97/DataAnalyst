```python
1.
missing_percentage = df['Postal Code'].isna().mean() * 100

print(f"Percentage of missing values in 'Postal Code': {missing_percentage:.2f}%")
```

```python
3.
customer_order_counts = df.groupby('Customer ID')['Order ID'].nunique()

# Filter customers who have more than one order
repeated_customers = customer_order_counts[customer_order_counts > 1]

# Get the number of customers with repeated orders
num_repeated_customers = repeated_customers.count()

# Display the frequency of repeat orders for these customers
repeat_order_frequencies = repeated_customers.value_counts().sort_index()

print(f"Number of customers with repeated orders: {num_repeated_customers}")
print("\nFrequency of repeat orders per customer:")
print(repeat_order_frequencies)
```

```python
4.
#  Count the number of unique orders for each customer
customer_order_counts = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
customer_order_counts.columns = ['Customer ID', 'Order Count']

#  Merge this data with customer segments
df_customers = df[['Customer ID', 'Segment']].drop_duplicates()
customer_segments = pd.merge(customer_order_counts, df_customers, on='Customer ID')

#  Identify repeat customers (Order Count > 1)
customer_segments['Is Repeat Customer'] = customer_segments['Order Count'] > 1

#  Calculate percentage of repeat customers by segment
repeat_customers_by_segment = customer_segments.groupby('Segment')['Is Repeat Customer'].mean() * 100

print("Percentage of repeat customers by segment:")
print(repeat_customers_by_segment)
```

```python
3.
# Converting dates to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format='mixed')

# Calculate the time difference
df['Time Taken'] = (df['Ship Date'] - df['Order Date']).dt.days

# Average time taken
avg_time_taken = df['Time Taken'].mean()
print(f"Average time taken between Order Date and Ship Date: {avg_time_taken:.2f} days")
```

```python
5.

# Convert Order Date to datetime if not already
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Resample orders by month
orders_by_month = df.resample('ME', on='Order Date').size()
print("Orders by Month:")
print(orders_by_month)

# Resample orders by quarter
orders_by_quarter = df.resample('QE', on='Order Date').size()
print("\nOrders by Quarter:")
print(orders_by_quarter)

# Resample orders by year
orders_by_year = df.resample('YE', on='Order Date').size()   
print("\nOrders by Year:")
print(orders_by_year)
```
